## TASK 2 KPIs

**Completion Rate: Percentage of users reaching the confirmation step.

Time Spent on Each Step: Average time users spend at each step in the process.

Error Rates: Cases where users navigate backward (from a later step to an earlier one).**

### Completion Rate:

```python
# List of required steps
required_steps = ['start', 'step_1', 'step_2', 'step_3', 'confirm']
# Group by client_id and visit_id and check the present process_step
steps_check = df_merged.groupby(['client_id', 'visit_id'])['process_step'].apply(set).reset_index()
# Check if all required steps are present
steps_check['all_steps_completed'] = steps_check['process_step'].apply(lambda x: set(required_steps).issubset(x))

completed_steps = steps_check[steps_check['all_steps_completed'] == True]

# Find the total number of these users
completed_users = completed_steps['client_id'].nunique()

# Find the total number of users in the data set
total_users = df_merged['client_id'].nunique()

# Calculate completion rate
completion_rate = (completed_users / total_users) * 100

# Print the result
print(f"Updated Completion Rate: {completion_rate}%")
```

### TIME SPENT ON EACH STEP WITHOUT OUTLIERS 

```python
# Convert date and time column to datetime format
df_merged["date_time"] = pd.to_datetime(df_merged["date_time"])

# Sort the data by visit_id, date_time, and client_id to ensure correct order
df_merged = df_merged.sort_values(by=["visit_id", "date_time", "client_id"], ascending=[True, True, True])

# Add a column to find the time of the next step
df_merged["next_step_time"] = df_merged.groupby("visit_id")["date_time"].shift(-1)

# Calculate time difference (in seconds)
df_merged["time_spent_seconds"] = (df_merged["next_step_time"] - df_merged["date_time"]).dt.total_seconds()

# Calculate IQR for time_spent_seconds
Q1 = df_merged["time_spent_seconds"].quantile(0.25)  # 25th percentile
Q3 = df_merged["time_spent_seconds"].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out rows with outlier values
df_filtered = df_merged[(df_merged["time_spent_seconds"] >= lower_bound) & (df_merged["time_spent_seconds"] <= upper_bound)]

# Recalculate the average time spent for each step after removing outliers
average_time_per_step = df_filtered.groupby("process_step")["time_spent_seconds"].mean()

print("Average Time Spent on Each Step (seconds) After Outlier Removal:")
print(average_time_per_step)
```

### ERROR RATE

```python
# We reorder the data to sort the steps within the same visit
df_merged = df_merged.sort_values(by=["visit_id", "date_time"])  # Veriyi sıralama

# Add a column to find the previous step
df_merged["previous_step"] = df_merged.groupby("visit_id")["process_step"].shift(1)

# Check for errors: A step returns to the previous stepsi
df_merged["is_error"] = df_merged.apply(
    lambda row: False
    if row["process_step"] == "confirm"
    else (row["process_step"] < row["previous_step"])
    if pd.notnull(row["previous_step"])
    else False,
    axis=1,
)

# Calculate error rate
total_steps = df_merged["is_error"].count()
total_errors = df_merged["is_error"].sum()
error_rate = (total_errors / total_steps) * 100

print(f"Total Steps: {total_steps}")
print(f"Total Errors: {total_errors}")
print(f"Error Rate: {error_rate:.2f}%")
```

## Task III: Hypothesis testing

Central Limit Theorem applies as population is large (> 50000). We can assume we have a normal distribution
two proportion z-test is most suited for proportion value and large sample

```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import ttest_ind
from scipy.stats import t
```

### 1-Completion rate

```python
#KPIS to be used
x_test = control_group_total
n_test = test_group_total
x_control = control_group_confirm
n_control = test_group_confirm
```

Null hypothesis: The test group does not have a higher completion rate (test <= control)

Alternative hypothesis: The test group has a higher completion rate (test > control)

```python
alpha = 0.05
#We accept to be wrong 5 % of the time 
```

T-Test

```python
# Calculate completion rates
p_test = x_test / n_test
p_control = x_control / n_control

# Observed difference
observed_diff = p_test - p_control

se_test = np.sqrt((p_test * (1 - p_test)) / n_test)
se_control = np.sqrt((p_control * (1 - p_control)) / n_control)
pooled_se = np.sqrt(se_test**2 + se_control**2)

# Degrees of freedom (conservative approximation)
df = min(n_test, n_control) - 1

mu_1 = 0  # Testing if p_test is greater than p_control (no difference in null)


t_stat_1 = (observed_diff - mu_1) / pooled_se
p_value_1 = t.sf(t_stat_1, df)  # One-sided t-test for Hypothesis 1

# Print results
print("Hypothesis 1: p_test > p_control")
print(f"Observed Difference: {observed_diff:.4f}")
print(f"T-Statistic: {t_stat_1:.4f}")
print(f"P-Value: {p_value_1}")
if p_value_1 < 0.05:
    print("The test group completion rate is significantly higher than the control group.")
else:
    print("The test group completion rate is not significantly higher than the control group.")
print("\n")
```

**Two proportion z-test**

```python

successes = np.array([x_test, x_control])
samples = np.array([n_test, n_control])

# Perform one-sided two-proportion z-test
z_stat_1, p_value_1 = proportions_ztest(successes, samples, alternative='larger')

print("Hypothesis 1: p_test > p_control")
print(f"Z-Statistic: {z_stat_1:.4f}")
print(f"P-Value: {p_value_1}")
if p_value_1 < 0.05:
    print("The test group completion rate is significantly higher than the control group.")
else:
    print("The test group completion rate is not significantly higher than the control group.")
print("\n")
```

### 2-Completion rate > 5%

H0 : Completion rate is =< 5% higher for test group (p_test - p_control <= 0.05)

H1 : Completion rate is > 5% higher for test group (p_test - p_control > 0.05)

Null hypothesis difference (mu)

5% de difference
mu=0.05

**T-Test**

```python
mu_2 = 0.05  # Testing if the difference is greater than 5%

t_stat_2 = (observed_diff - mu_2) / pooled_se

p_value_2 = t.sf(t_stat_2, df)  # One-sided t-test for Hypothesis 2

print("Hypothesis 2: p_test - p_control > 0.05")
print(f"Observed Difference: {observed_diff:.4f}")
print(f"T-Statistic: {t_stat_2:.4f}")
print(f"P-Value: {p_value_2}")
if p_value_2 < 0.05:
    print("The test group completion rate is significantly at least 5% higher than the control group.")
else:
    print("The test group completion rate is not significantly at least 5% higher than the control group.")
```

**Two proportion z-test**

```python
mu = 0.05

# Perform one-sided two-proportion z-test with null difference
z_stat_2, p_value_2 = proportions_ztest(successes, samples, value=mu, alternative='larger')

print("Hypothesis 2: p_test - p_control > 0.05")
print(f"Z-Statistic: {z_stat_2:.4f}")
print(f"P-Value: {p_value_2:.4f}")
if p_value_2 < 0.05:
    print("The test group completion rate is significantly at least 5% higher than the control group.")
else:
    print("The test group completion rate is not significantly at least 5% higher than the control group.")
```

### 3-Old and new client difference

Two samples T-Test

```python

alpha = 0.1
from scipy.stats import ttest_ind
df_test_tenure = df_unique_clients[df_unique_clients["Variation"]=="Test"]["clnt_tenure_yr"]
df_control_tenure = df_unique_clients[df_unique_clients["Variation"]=="Control"]["clnt_tenure_yr"]
t_stat, p_value = ttest_ind(df_test_tenure, df_control_tenure, equal_var=False)
#two tailed test is fine as we just want =
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.5f}")

if p_value > alpha:
    print("Fail to reject the null hypothesis: No significant evidence of a difference in tenure between the test and control groups.")
else:
    print("Reject the null hypothesis: There is a significant difference in the tenure between the test and control groups.")
```
